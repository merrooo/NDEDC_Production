```python
import threading # this for watching dog
import time

class Car:

    def __init__(self):
        self.ignition_key = False
        self.sensor_value = 0
        self.dashboard = "OFF"

    # 🚗 Engine start
    def startEngine(self):
        self.ignition_key = True

        # start watchdog immediately
        threading.Thread(target=self.watchdog, daemon=True).start()

    # 🐕 WATCHDOG (runs always)
    def watchdog(self):

        while self.ignition_key:

            # NORMAL
            if 0 <= self.sensor_value <= 3:
                self.dashboard = "NORMAL"

            # WARNING
            elif 4 <= self.sensor_value <= 6:
                self.dashboard = "WARNING"
                print("🟡 WARNING:", self.sensor_value)

            # CRITICAL
            elif self.sensor_value > 6:
                self.dashboard = "CRITICAL"
                print("🔴 OVERHEAT! EMERGENCY STOP")
                self.stopEngine()
                break

            time.sleep(0.5)

    # 🚗 Running engine (sensor input simulation)
    def runEngine(self):

        count = 0

        while self.ignition_key :

            self.sensor_value = int(input("Sensor Reading :- "))

            print(f"dashboard: {self.dashboard}")

            count += 1

    # 🛑 Stop engine
    def stopEngine(self):
        self.ignition_key = False
        print("Engine STOPPED")


# ================= MAIN =================
if __name__ == "__main__":

    car = Car()

    car.startEngine()
    car.runEngine()
    car.stopEngine()

```
